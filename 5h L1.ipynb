```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Lasso
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)
model_lr = LinearRegression()
model_l1 = Lasso(alpha=10)
model_lr.fit(X, y)
model_l1.fit(X, y)
xx = np.linspace(X.min()-1, X.max()+1, 100).reshape(-1,1)
y_pred_lr = model_lr.predict(xx)
y_pred_l1 = model_l1.predict(xx)
plt.scatter(X, y, c='black')
plt.plot(xx, y_pred_lr, label='Linear Regression', color='blue')
plt.plot(xx, y_pred_l1, label='L1 Regularization (Lasso)', color='red')
plt.title('Effect of L1 Regularization')
plt.legend()
plt.show()
```
