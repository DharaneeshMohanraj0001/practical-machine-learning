```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
X, y = make_classification(
 n_samples=300,
 n_features=6, 
 n_informative=4, 
 n_redundant=1, 
 n_clusters_per_class=2,
 flip_y=0.15, 
 random_state=42
)
model = LogisticRegression(max_iter=1000)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=kf, scoring='roc_auc')
print("AUC Scores for each fold:", scores)
print("Average AUC Score:", np.mean(scores))
plt.figure(figsize=(8,5))
plt.plot(range(1, 6), scores, marker='o', linestyle='-', linewidth=2, color='orange')
plt.title("5-Fold Cross-Validation ROC-AUC (Noisy Dataset)")
plt.xlabel("Fold Number")
plt.ylabel("AUC Score")
plt.ylim(0.6, 1.0)
plt.grid(True)
plt.show()
```
