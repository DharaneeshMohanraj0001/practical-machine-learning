```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = {
 "Size_sqft": [850, 900, 1200, 1500, 1800, 2000, 2300, 2500, 2700, 3000],
 "Rooms": [2, 2, 3, 3, 4, 4, 4, 5, 5, 6],
 "Price": [100, 115, 150, 180, 210, 240, 270, 300, 330, 360]
}
df = pd.DataFrame(data)
X = df[["Size_sqft", "Rooms"]]
y = df["Price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svr = SVR(kernel='rbf') 
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print(" R² Score:", r2)
print(" Actual:", list(y_test.values))
print(" Predicted:", list(y_pred))
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', label="Predicted vs Actual")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("SVM Regression - Actual vs Predicted")
plt.legend()
plt.show()
```
