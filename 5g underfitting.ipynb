```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X, y = make_moons(n_samples=200, noise=0.25, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model_underfit = DecisionTreeClassifier(max_depth=1)
model_underfit.fit(X_train, y_train)
xx, yy = np.meshgrid(np.linspace(X[:,0].min()-0.5, X[:,0].max()+0.5, 300),np.linspace(X[:,1].min()-0.5, 
X[:,1].max()+0.5, 300))
Z = model_underfit.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.figure(figsize=(6,5))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X_train[:,0], X_train[:,1], c=y_train)
plt.title(f'Underfitted Model\nTrain:{accuracy_score(y_train, model_underfit.predict(X_train)):.2f} 
Test:{accuracy_score(y_test, model_underfit.predict(X_test)):.2f}')
plt.show()
```
