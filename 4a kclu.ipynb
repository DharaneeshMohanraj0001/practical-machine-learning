```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
iris = load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
np.random.seed(42)
noise = np.random.normal(0, 0.1, X_scaled.shape)
X_noisy = X_scaled + noise
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_noisy)
plt.figure(figsize=(7,5))
plt.scatter(X_noisy[:, 0], X_noisy[:, 1], c=labels, cmap='rainbow', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', s=200, marker='X', 
label='Centroids')
plt.title("K-Means Clustering on Noisy Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
