```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt
import numpy as np
iris = load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
np.random.seed(42)
noise = np.random.normal(0, 0.1, X_scaled.shape)
X_noisy = X_scaled + noise
linkage_matrix = linkage(X_noisy, method='ward')
labels = fcluster(linkage_matrix, t=3, criterion='maxclust')
plt.figure(figsize=(7,5))
plt.scatter(X_noisy[:, 0], X_noisy[:, 1], c=labels, cmap='rainbow', s=50)
plt.title("Hierarchical Clustering on Noisy Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
