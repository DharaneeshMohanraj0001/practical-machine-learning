```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)
model_lr = LinearRegression()
model_l2 = Ridge(alpha=10)
model_lr.fit(X, y)
model_l2.fit(X, y)
xx = np.linspace(X.min()-1, X.max()+1, 100).reshape(-1,1)
y_pred_lr = model_lr.predict(xx)
y_pred_l2 = model_l2.predict(xx)
plt.scatter(X, y, c='black')
plt.plot(xx, y_pred_lr, label='Linear Regression', color='blue')
plt.plot(xx, y_pred_l2, label='L2 Regularization (Ridge)', color='green')
plt.title('Effect of L2 Regularization')
plt.legend()
plt.show()
```
