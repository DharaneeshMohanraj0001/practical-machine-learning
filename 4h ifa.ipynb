```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
np.random.seed(42)
normal_data = 0.3 * np.random.randn(100, 2)
normal_data = np.r_[normal_data + 2, normal_data - 2]
anomalies = np.random.uniform(low=-6, high=6, size=(10, 2))
data = np.r_[normal_data, anomalies]
df = pd.DataFrame(data, columns=["Feature 1", "Feature 2"])
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(df)
df['Anomaly'] = model.predict(df)
normal = df[df['Anomaly'] == 1]
anomaly = df[df['Anomaly'] == -1]
plt.figure(figsize=(7,5))
plt.scatter(normal['Feature 1'], normal['Feature 2'], c='blue', label='Normal Data', s=40)
plt.scatter(anomaly['Feature 1'], anomaly['Feature 2'], c='red', label='Anomaly', s=60)
plt.title("Anomaly Detection using Isolation Forest")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
